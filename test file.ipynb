```python
import os
from PIL import Image
import matplotlib.pyplot as plt

image_dir = 'dataset\\training_data\\images'  # or test_data
image_paths = [os.path.join(image_dir, img) for img in os.listdir(image_dir) if img.endswith('.png')]

# Preview an image
img = Image.open(image_paths[0])
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
import pytesseract
import time
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
def ocr_with_tesseract(image_path):
    img = Image.open(image_path)
    start = time.time()
    text = pytesseract.image_to_string(img)
    elapsed = time.time() - start
    return text, elapsed
ocr_with_tesseract(image_paths[0])
```

```python
import easyocr

reader = easyocr.Reader(['en'])

def ocr_with_easyocr(image_path):
    start = time.time()
    result = reader.readtext(image_path)
    text = ' '.join([res[1] for res in result])
    elapsed = time.time() - start
    return text, elapsed
```

```python
import keras_ocr

pipeline = keras_ocr.pipeline.Pipeline()

def ocr_with_kerasocr(image_path):
    start = time.time()
    img = keras_ocr.tools.read(image_path)
    prediction = pipeline.recognize([img])[0]
    text = ' '.join([word[0] for word in prediction])
    elapsed = time.time() - start
    return text, elapsed
```

```python
from paddleocr import PaddleOCR

ocr_paddle = PaddleOCR(use_angle_cls=True, lang='en')

def ocr_with_paddleocr(image_path):
    start = time.time()
    result = ocr_paddle.ocr(image_path)[0]
    text = ' '.join([line[1][0] for line in result])
    elapsed = time.time() - start
    return text, elapsed
```

```python
results = []

for tool, func in [
    ('Tesseract (pytesseract)', ocr_with_tesseract),
    ('EasyOCR', ocr_with_easyocr),
    #('Keras-OCR', ocr_with_kerasocr),
    ('PaddleOCR', ocr_with_paddleocr)
]:
    text, t = func(image_paths[0])
    results.append({
        'Tool': tool,
        'Accuracy (%)': '-',  
        'Time per Image (s)': round(t, 2),
        'Sample Output': text[:300] + '...'
    })

import pandas as pd
pd.DataFrame(results)
```

```python
from flask import Flask, request, jsonify
from PIL import Image
import pytesseract
import os

# If Tesseract is not in PATH, set it manually
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

app = Flask(__name__)

@app.route('/extract-text', methods=['POST'])
def extract_text():
    if 'image' not in request.files:
        return jsonify({"error": "No image file provided"}), 400
    
    image_file = request.files['image']
    image = Image.open(image_file.stream)
    extracted_text = pytesseract.image_to_string(image)
    
    return jsonify({"text": extracted_text})

if __name__ == '__main__':
    app.run(debug=True)
```
